## Taller Redes Neuronales Clasificaci√≥n

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el dataset de "human activity" un DataFrame de pandas
dfdata = pd.read_csv("mhealth_raw_data.csv")
dfdata.head()
```

```python
# Imprimir la cola
dfdata.tail()
```

```python
# Verificar la cantidad de datos por cada columna
dfdata.count()
```

```python
# Revisar la estad√≠stica descriptiva del dataset
dfdata.describe()
```

```python
# Crear una lista con los labels de cada clase, excepto la clase 0: None
labels = ['Standing still','Sitting and relaxing','Lying down','Walking'
         ,'Climbing stairs','Waist bends forward','Frontal elevation of arms'
         , 'Knees bending', 'Cycling','Jogging','Running','Jump']

labels
```

```python
# Graficar el n√∫mero de datos asociado a cada tipo de actividad para verificar si el dataset est√° o no balanceado
sns.countplot(x='Activity',data=dfdata)
plt.xticks(x = dfdata['Activity'])
plt.show()
```

```python
# Verificar el n√∫mero de elementos de la actividad con menos datos
dfdata[dfdata['Activity']==12].count()
```

```python
# Remuestrear el dataset para balancear clases, usar 10100 datos para cada clase
from sklearn.utils import resample

dfdata2 = pd.DataFrame()
for i in range(12):
    df_aux = dfdata[dfdata.Activity==i+1]
    df_aux = resample(df_aux, n_samples=10100, random_state=12)
    dfdata2 = pd.concat([dfdata2,df_aux])
```

```python
# Forzar a que las etiquetas de las actividades inicien en 0
dfdata2['Activity'] = dfdata2['Activity']-1
```

```python
# Verificar nuevamente el n√∫mero de datos por actividad
sns.countplot(x='Activity',data=dfdata2)
plt.xticks(x = dfdata2['Activity'])
plt.show()

```

```python
dfdata2.head()
```

```python
# Construir las variables independientes X y la dependiente Y a partir del dataset balanceado
X = dfdata2.drop(['subject','Activity'],axis = 1)
Y = dfdata2['Activity']

# Convertir esos dataframes en array numpy
X = np.asarray(X)
Y = np.asarray(Y)

print(X.shape,Y.shape)
```

```python
# Dividir el dataset en datos de entrenamiento (75%) y prueba (25%)
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 0.25, random_state = 42)

xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
# Usar un encoder para convertir la forma de la salida y
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore')
enc.fit(np.asarray(ytrain).reshape(-1,1))

y_train = enc.transform(np.asarray(ytrain).reshape(-1,1)).toarray()
y_test = enc.transform(np.asarray(ytest).reshape(-1,1)).toarray()
# y puede tomar valores de 0-10, convertimos a y en una matriz de 11 columnas, en donde la columna que determina la clase
# de cada muestra vale 1 y los dem√°s valores son 0 
```

# Red Neuronal

```python
# Crear un modelo de red neuronal totalmente conectada con las siguientes caracter√≠sticas
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

num_clases = 12

model = Sequential()
model.add(Dense(1024, input_shape=(xtrain.shape[1],),activation='relu'))
model.add(Dropout(0.05))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.05))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.05))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.05))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.05))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.05))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.05))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.05))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.05))
model.add(Dense(32,activation='relu'))
model.add(Dropout(0.1))

model.add(Dense(num_clases,activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer = 'adagrad',metrics=['accuracy'])

model.summary()
# 10 Capas densas que inicien con 1024 neuronas y disminuyan progresivamente hasta la capa de salida
# Despu√©s de cada capa densa un dropout de 5%
# Usar funciones de activaci√≥n relu para las capas ocultas y softmax para las capas de salida
# Usar categorical cross entropy como funci√≥n de error y optimizado adagrad
# Para el entrenamiento calcular los steps dado un batch_size = 16
# Imprimir el summary del modelo
```

```python
# Entrenar el modelo durante 50 √©pocas con un batch_size de 16 y guardar el hist√≥rico
# Usar una divisi√≥n de datos para validaci√≥n de 15%
epochsk =50


history = model.fit(xtrain,y_train, epochs = epochsk, batch_size = 16, verbose =1, validation_split = 0.15)
```

```python
# Guardar el modelo en el disco con nombre TallerRedesNeuronales.h5
model.save('TallerRedesNeuronales.h5')# 
```

```python
# Graficar loss y accuracy de entrenamiento y validaci√≥n
plt.subplot(2,1,1)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['train_loss','val loss'])

plt.subplot(2,1,2)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['train acc','val acc'])
```

```python
# Estimar el desempe√±o del modelo usando la matriz de confusi√≥n y las m√©tricas precisi√≥n, recall, accuracy y f1score
#calculamos la matriz de confusion 
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix
from sklearn import metrics

ypred= model.predict(xtest)
matriz=confusion_matrix(ytest, np.argmax(ypred, axis= 1))
plot_confusion_matrix(conf_mat=matriz, figsize=(6,6), show_normed =False)
plt.tight_layout()

print (metrics.classification_report(ytest, np.argmax(ypred, axis= 1), digits=4))
```
